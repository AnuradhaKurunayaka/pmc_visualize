```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('sampleset (Autosaved) (1)excel.xlsx')
```

```python
df.head()
```

## No of Diseases in DataSet

```python
diseases = list(df['Disease'].unique())
print(diseases)
print()
print("No of Diseases = ",len(diseases))
```

## Age Distribution based on Disease

```python
plt_list =[]
plt.figure(figsize=(12,10))
for disease in diseases:
    disease_df = df.loc[df['Disease'] == disease]
    sns.distplot(disease_df['Age'], label=disease,bins=5)
    plt.legend()
```

## Age VS Gender

```python
sns.boxplot(x='Age', y='Gender', data=df)
plt.show()
```

## Gender Count Based on Disease

```python
plt_list =[]
for disease in diseases:
    disease_df = df.loc[df['Disease'] == disease]
    sns.countplot(disease_df['Gender'], label = disease);
    plt.legend()
    plt.show()
```

```python
plt_list =[]
for disease in diseases:
    print(disease)
    disease_df = df.loc[df['Disease'] == disease]
    sns.boxplot(x='Age', y='Gender', data=disease_df)
    plt.show()
```
