```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('data_set/Book2.xlsx')
```

```python
df2 = pd.read_excel('data_set/Book2-2 (1).xlsx')
```

```python
df2.info()
```

```python
merged = pd.concat([df1, df2], ignore_index=True)
```

```python
merged
```

```python
merged['Disease'].unique()
```

```python
Gender = pd.get_dummies(merged['Gender'],drop_first=True)
```

```python
merged = pd.concat([merged,Gender],axis=1)
```

```python
merged.drop(['Gender'],axis=1,inplace=True)
```

```python
merged.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = merged.drop('Disease',axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,merged['Disease'],
                                                    test_size=0.30,random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
